```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
def analyze_sentiment(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)
    # We'll consider the compound score for the overall sentiment
    return sentiment_scores['compound']

df['Sentiment'] = df['Headlines'].apply(analyze_sentiment)
```

```python
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

df['NER_results'] = df['Headlines'].apply(lambda x: {ent.label_: ent.text for ent in nlp(x).ents})
```

```python
df
```

```python
df['KeyWords'] = df['NER_results'].apply(lambda x: [value for key, value in x.items() if key in ['GPE', 'LOC', 'PERSON', 'ORG']])
```

```python


# Initialize an empty dictionary to store the counts
word_counts = {}

# Iterate over the 'NER_results' column
for ner_result in df['NER_results']:
    # Check if the dictionary contains the desired keys
    for key, value in ner_result.items():
        if key in ['GPE', 'LOC', 'PERSON', 'ORG']:
            # Increment the count for each word
            word_counts[value] = word_counts.get(value, 0) + 1

# Display the word counts
print(word_counts)
```

```python
df_word_counts = pd.DataFrame(list(word_counts.items()), columns=['word', 'frequency'])

# Display the DataFrame
print(df_word_counts)
```
